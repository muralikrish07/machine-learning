```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('C:/Users/dheer/Desktop/haberman (1).csv')
df.head()
```

```python
#df['status'].value_counts()
df.describe(include='all')
```

```python
df.info()
```

```python
df['survived_status']=df['status'].map({1:'Survived',2:'Dead'})
df.tail()
```

```python
df['survived_status'].value_counts().plot(kind='bar')
```

```python
plt.scatter(df['age'],df['axil_nodes'],c='b')
plt.title("age vs axil_nodes")
plt.xlabel("age")
plt.ylabel("axil_node")
plt.show()
```

```python
plt.scatter(df['axil_nodes'],df['operation_year'],c='y')
plt.title("age vs axil_nodes")
plt.xlabel("age")
plt.ylabel("axil_node")
plt.show()
```

```python
plt.close();
sns.set_style('whitegrid');
sns.pairplot(df, hue = 'survived_status', size = 4)
plt.show()
```

```python
sns.FacetGrid(df,hue='survived_status',size=6)\
    .map(sns.distplot,'age')\
    .add_legend()
plt.show()
```

```python
sns.FacetGrid(df,hue='survived_status',size=6)\
    .map(sns.distplot,'operation_year')\
    .add_legend()
plt.show()
```

```python
sns.boxplot(x='survived_status', y = 'axil_nodes', data=df)
plt.show()
```

```python
sns.violinplot(x='survived_status', y = 'axil_nodes', data=df,size=9)
plt.show()
```

```python
survived_sta=df['status'].values
```

```python
fig = plt.figure(figsize=plt.figaspect(.5))

ax1 = plt.subplot(121)
sns.kdeplot(survived_sta, shade=True,ax=ax1)
plt.xlabel('survival_status')
plt.title("PDF")
ax2 = plt.subplot(122)
sns.kdeplot(survived_sta, shade=True, cumulative=True,ax=ax2)
plt.xlabel('survived_status')
plt.title('CDF')

plt.show()



plt.show()
```
